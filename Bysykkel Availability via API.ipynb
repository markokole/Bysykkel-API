```python
client_ident = '771f3bd92377b2d65e8d46d2fb4fef64'
```

```python
import requests
import json
import pandas as pd
from pandas.io.json import json_normalize
```

```python
root_url = 'https://oslobysykkel.no/api/v1/'
```

```python
def get_response(client_ident, endpoint):
    """
    Fullfører API kall
    Inputs:
    - client_ident: client-identifier
    - endpoint: endpoint i API url
    Return: 
    - status kode om API kall gikk gjennom (kode 200) eller ikke - type int
    - text levert av API kall - type str
    """
    try:
        headers = {'Client-Identifier': client_ident, 'Content-Type': 'application/json'}
        response= requests.get(url=root_url+endpoint, headers=headers)
    except:
        return 0, "get_response: Feilmelding ikke relatert til API direkte."
    return response.status_code, response.text
```

```python
status_code_stations, response_stations = get_response(client_ident, "stations")

if (status_code_stations == 200):
    # hent data om "stations"
    stations_tbl = pd.DataFrame.from_dict(json_normalize(json.loads(response_stations)["stations"]), orient='columns')
    
    status_code_availability, response_availability = get_response(client_ident, "stations/availability")
    if (status_code_availability == 200):
        # hent data om "stations availability"
        availability_tbl = pd.DataFrame.from_dict(json_normalize(json.loads(response_availability)["stations"]), orient='columns')

        # gjør join på de to DataFramene for å få status om "stations"
        stations_status = stations_tbl.join(availability_tbl.set_index("id"), on="id")[["title", "availability.bikes", "availability.locks"]]
        display(stations_status) # vis data
    else:        
        print("Status ved henting data om 'stations availability': %s" % (response_availability))
else:
    print("Status ved henting data om 'stations': %s" % (response_stations))
    
```
